```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Case 1

age_x=[25,26,27,29,33,32,35,28,35,39]
devop=[3525,3326,3627,3129,3633,3032,3115,3628,3435,3009]
javadop=[4525,5326,5627,5129,5633,5032,6115,6628,6435,6009]
age_x.sort()
devop.sort()
javadop.sort()
```

```python
plt.plot(age_x,devop,label="All devep")
plt.plot(age_x,javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()

```

```python
# Case 2

age_x=np.random.normal(35,5,10)
devop=np.random.normal(35000,10000,10)
javadop=np.random.normal(50000,5000,10)
age_x.sort()
devop.sort()
javadop.sort()

plt.plot(age_x,devop,label="All devep")
plt.plot(age_x,javadop,label="All Java ")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
plt.legend()
```
